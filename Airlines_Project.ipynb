# **Q.4 Does price vary with Airlines?**

```python
df.head()
```

```python
## Grouping the airlines and checking their mean prices.

df.groupby('airline')['price'].mean()
```

```python
## Drawing a categorical plot showing the mean ticket price for each Airline. We are using Seaborn because it Automatically groups data
fig, axes = plt.subplots(1, 2, figsize=(14,5))
sorted_df = df.sort_values(by='price', ascending=True)
sns.barplot(x='airline', y='price', data=df, palette='rocket', hue = 'class', ax=axes[0])
axes[0].set_title("Average Price per Airline (Seaborn)")
axes[0].tick_params(axis='x', rotation=45)

plt.subplot(1,2,2)
df.groupby('airline')['price'].mean().plot(kind = 'bar', color=['#FFA500', '#FF8C00', '#FF7F50', '#FF6347', '#FF4500', '#FFDAB9'])
plt.title('Average Price per Airline (Matplotlib)')
plt.xlabel('Airline')
plt.ylabel('Average Price')


plt.show()
```

# **Q5. Does** Ticket price change based on the departure time and arrival time

```python
## Checking the mean Ticket Price based on the Departure time

df.groupby('departure_time')['price'].mean()
```

```python
## Checking the mean Ticket Price based on the Arrival time

df.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x ='departure_time', y = 'price', kind = 'bar', palette='rocket', data = df)
plt.title('Average Price as per Departure Time')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.catplot(x ='arrival_time', y = 'price', kind = 'bar', palette='rocket', data = df)
plt.title('Average Price as per Arrival Time')
plt.xticks(rotation=45)
plt.show()
```

Creating a Relational Plot.

```python
sns.relplot( x ='arrival_time', y = 'price', kind = 'line', palette='rocket', col = 'departure_time', hue='departure_time', data = df)
plt.xticks(rotation=45)
plt.show()
```
